### Ejercicio de Escalabilidad

Suponga que un programa tiene tiempo Tserial = n (microsegundos) y Tparalelo = n/p + log2(p).  Escriba un programa que evalúe el Speedup y la Eficiencia de un problema con Dominio creciente n (10, 20, 30, … 320) y Núcleos p (1, 2, 4, 8, 16, 32).

Determine si el problema es Escalable y en qué grado.  Si es necesario, calcule el valor del factor k de crecimiento de p.

Entrege su documento como un informe de 2 hojas máximo, incluyendo al menos una gráfica.  Adjunte el código fuente de su  - programa o script.  Puede trabajar en cualquier lenguaje (C/C++, Python, Java, R) o programa de cálculo (Mathlab, Mathematica, Octave)

```python
import math
import matplotlib.pyplot as plt

class Escalabilidad:
    def __init__(self,p):
        '''p: numero de nucleos
           speedUp: S = Tserial/Tparalelo
           eficiencia: E = S/p
        '''
        self.p = p
        self.S = []
        self.E = []
        
    def tParalelo(self,n,p): #Tiempo paralelo = n/p + log2(p)
        tiempo = n/p + math.log2(p)
        return tiempo
    
    def speedUpYEficiencia(self):
        for n in range(10,320,10):
            tiempoSerial = n
            speedUpLineal = tiempoSerial/self.tParalelo(n,self.p)
            self.S.append(speedUpLineal)
            
        for s in self.S:
            efficiency = s/self.p
            self.E.append(efficiency)
            
        return self.S, self.E
    
    
nucleos = [1, 2, 4, 8, 16, 32]

esc = Escalabilidad(1)
s1,e1 = esc.speedUpYEficiencia()
#print (s1,e1)
esc2 = Escalabilidad(2)
s2,e2 = esc2.speedUpYEficiencia()
#print (s2,e2)
esc3 = Escalabilidad(4)
s3,e3 = esc3.speedUpYEficiencia()
#print (s3,e3)
esc4 = Escalabilidad(8)
s4,e4 = esc4.speedUpYEficiencia()
#print (s4,e4)
esc5 = Escalabilidad(16)
s5,e5 = esc5.speedUpYEficiencia()
#print (s5,e5)
esc6 = Escalabilidad(32)
s6,e6 = esc6.speedUpYEficiencia()
#print (s6,e6)
```

```python
plt.figure(figsize=(8,4))
plt.title("Speed up para 2 nucleos")
plt.plot(s2)
plt.ylabel("Nucleos")
plt.show()

plt.figure(figsize=(8,4))
plt.title("Speed up para 4 nucleos")
plt.ylabel("Nucleos")
plt.plot(s3)
plt.show()
```
